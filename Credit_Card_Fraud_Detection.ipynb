```python
import imblearn
from imblearn.over_sampling import SMOTE
sm=SMOTE(random_state=42) 
sm_x,sm_y=sm.fit_resample(x,y)
sm_x=pd.DataFrame(sm_x)
sm_y=pd.DataFrame(sm_y)
df_sm=pd.concat([sm_x,sm_y],axis=1)
df_sm['Class'].value_counts().plot(kind='bar')
```

```python
df_fraud=df_sm[df_sm['Class']==1]
df_normal=df_sm[df_sm['Class']==0]
df_fraud=df_fraud.sample(frac=0.5, replace=True, random_state=1)
df_normal=df_normal.sample(frac=0.5, replace=True, random_state=1)
df_fraud=df_fraud.reset_index()
df_normal=df_normal.reset_index()
df_sm=pd.concat([df_fraud,df_normal],axis=0)
df_sm.shape
```

```python
x=df_sm.iloc[:,0:-1]
y=df_sm.iloc[:,-1]
col=[i for i in x.columns]
```

```python
def del_outlier(df, feature):
    for f in feature:
        # lower quartile
        q1 = np.percentile(df[f], 25)

        # upper quartile
        q3 = np.percentile(df[f], 75)

        # interquartile range
        iqr = q3 - q1

        # with coefficient
        with_coef = 1.5 * iqr

        # detect outlier(s)
        lower_rule = q1 - with_coef
        upper_rule = q3 + with_coef

        outlier_variable = df[(df[f] < lower_rule) | (df[f] > upper_rule)].index
        df[f].loc[outlier_variable]=np.nan
    return df
```

```python
x=del_outlier(x,col)
x.isnull().sum()/df.shape[0]
```

```python
x=x.fillna(value=x.median())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
new_x=scaler.fit_transform(x)
x_ss=pd.DataFrame(new_x,columns=x.columns)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=9).fit_transform(x_ss)
x_pca=pd.DataFrame(pca)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.3, random_state=21)
```

```python
from sklearn.svm import SVC
svm_model = SVC( C=7,kernel='rbf',probability=True).fit(X_train, y_train)
#Grid Search

#svm_params = {"C": np.arange(1,10), "kernel":["linear", "rbf"]}
#svm_cv_model = GridSearchCV(svm_model, svm_params, cv=7, n_jobs=-1, verbose=7).fit(X_train, y_train)
#svm_cv_model.best_score_
#best_params = svm_cv_model.best_params_
#print(best_params)
#svm = SVC(C = best_params['C'], kernel=best_params['kernel'], probability=True).fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score
y_pred_svm = svm_model.predict(X_test)
accuracy_score(y_test, y_pred_svm)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(svm_model, X_test, y_test, cv=21).mean()
print(classification_report(y_test, y_pred_svm))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, oob_score=True, n_jobs=4)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
cross_val_score(rf, X_test, y_test, cv=21).mean()
print(classification_report(y_test, y_pred_rf))
```

```python
accuracy_score(y_test, y_pred_rf)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=50,
                    solver='sgd', verbose=10, momentum=0.9,
                    learning_rate='adaptive', learning_rate_init=0.1)
mlp.fit(X_train, y_train)
y_pred_ml = mlp.predict(X_test)
cross_val_score(mlp, X_test, y_test, cv=21).mean()
print(classification_report(y_test, y_pred_ml))
```

```python
accuracy_score(y_test, y_pred_ml)
```
